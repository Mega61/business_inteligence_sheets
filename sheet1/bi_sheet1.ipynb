### Exercise 1 (2 Points) Create two lists $L1$ and $L2$. The first with letters a, b, c, .., h and the second with numbers from 1 to 8  

1. 
    - a) Get familiar with the function zip and what it does. Print the 8 tuples ($L1_{i}$, $L2_{i}$), where $L1_{i}$ and $L2_{i}$ are the $i$-th entries of the respective lists. Use at most 4 lines.


```python
L1 = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
L2 = [1, 2, 3, 4, 5, 6, 7, 8]
x = zip(L1, L2)
print(list(x))
```

```python
for index, value in enumerate(L1, 1):
    for index2, value2 in enumerate(L2, 1):
        if(index >= index2):
            print(value, index2)
```

2. Create a 3×2 plot grid (use subplots of matplotlib) and insert the 6 2D-scatter plots
of the iris-dataset for every pair of the four numerical attributes. (Command: matplotlib.pyplot.scatter ). Use different colors for the different species and create a legend
(Command: matplotlib.pyplot.legend).  
Hint: Use the groupby function of the dataframe to group by class, and use itertools to
compute the combinations of attributes. This exercise can be solved in less than 10 lines
in total!!


```python
import matplotlib.pyplot as plt

species = iris_df.groupby(['species'])
fig, ax = plt.subplots(3, 2)
fig.suptitle('2D-Scatter plots')
fig.tight_layout()
for name, grouped in species:
    ax[0, 0].scatter(grouped.sepal_length,
                     grouped.sepal_width, label=name, marker='.')
    ax[0, 1].scatter(grouped.sepal_length, grouped.petal_length, marker='.')
    ax[1, 0].scatter(grouped.sepal_length, grouped.petal_width, marker='.')
    ax[1, 1].scatter(grouped.sepal_width, grouped.petal_length, marker='.')
    ax[2, 0].scatter(grouped.sepal_width, grouped.petal_width, marker='.')
    ax[2, 1].scatter(grouped.petal_length, grouped.petal_width, marker='.')
fig.legend()
```

5. Create a figure showing the joint empirical cumulative distribution of sepal_length and
sepal_width (in a 3D plot, use mesh grid or surface).  
Hint: Check Equation (2.21) in the book to recall the definition of this distribution.

```python
from matplotlib import pyplot

#Xi is iris database data
xi = iris_df[['sepal_length', 'sepal_width']].to_numpy()
n = xi[0].size
#Yi is the actual distribution
yi = np.arange(1, n+1)/n
#Zi is the max data range from the sepal length and width
zi = np.array([np.arange(0,10,0.067),np.arange(0,10,0.067)]).T

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(xi, yi, zi, cmap='viridis')
plt.show()
```
